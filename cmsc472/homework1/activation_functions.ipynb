```python
import matplotlib.pyplot as plt 
import numpy as np 
```

```python
x = np.linspace(-5, 5, 1000) 
y = [0 if z < 0 else z for z in x]
d = [0 if z < 0 else 1 for z in x]
  
plt.plot(x, y) 
plt.plot(x, d)
plt.xlabel("x") 
plt.title("Relu")
plt.ylabel("y") 
ax = plt.gca()
ax.set_ylim([-10, 10])
plt.legend(["Relu(x)", "Derivative"])
plt.show()
```

```python
x = np.linspace(-5, 5, 1000) 
y = (np.exp(x) - np.exp(-x))/(np.exp(x) + np.exp(-x)) 
d = 1 - ((np.exp(x) - np.exp(-x))/(np.exp(x) + np.exp(-x))) ** 2
  
plt.plot(x, y) 
plt.plot(x, d) 
plt.xlabel("x") 
plt.title("Tanh")
plt.ylabel("y") 
ax = plt.gca()
ax.set_ylim([-10, 10])
plt.legend(["Tanh(x)", "Derivative"])
plt.show() 
```

```python
x = np.linspace(-5, 5, 1000) 
y = 1/(1 + np.exp(-x)) 
d= (np.exp(-x)) / ((1 + np.exp(-x)) ** 2)
  
plt.plot(x, y) 
plt.plot(x, d)
plt.xlabel("x") 
plt.title("Sigmoid")
plt.ylabel("y") 
ax = plt.gca()
ax.set_ylim([-10,10])
plt.legend(["Sigmoid(x)", "Derivative"])
plt.show() 
```

```python
x = np.linspace(-5, 5, 1000) 
y = [0.01 * z if z < 0 else z for z in x]
d = [0.01 if z < 0 else 1 for z in x]
  
plt.plot(x, y) 
plt.plot(x, d)
plt.xlabel("x") 
plt.title("Leaky Relu")
plt.ylabel("y") 
ax = plt.gca()
ax.set_ylim([-10, 10])
plt.legend(["Leaky Relu(x)", "Derivative"])
plt.show() 
```

```python
x = np.linspace(-5, 5, 1000) 
y = [z if z >= 0 else 0.03 * (np.exp(z) + 1) for z in x]
d = [0.03 * np.exp(z) if z < 0 else 1 for z in x]
  
plt.plot(x, y) 
plt.plot(x, d)
plt.title("Elu")
plt.xlabel("x") 
plt.ylabel("y") 
ax = plt.gca()
plt.legend(["Elu(x)", "Derivative"])
ax.set_ylim([-10, 10])
plt.show() 
```

```python
x = np.linspace(-5, 5, 1000) 
y = [np.sin(z)/z for z in x]
d = [np.cos(z) / z - np.sin(z) / (z **2) for z in x]
  
plt.plot(x, y) 
plt.plot(x, d)
plt.title("Sinc")
plt.xlabel("x") 
plt.ylabel("y") 
ax = plt.gca()
ax.set_ylim([-10, 10])
plt.legend(["Sinc(x)", "Derivative"])
plt.show() 
```
